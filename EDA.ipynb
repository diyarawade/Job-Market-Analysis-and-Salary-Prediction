```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/Diya/Downloads/cleaned_salaries.csv")
df.head()
```

```python
plt.figure()
plt.hist(df['salary_in_usd'])
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title("Salary vs Experience")
plt.show()
```

```python
plt.figure()
df['job_title'].value_counts().head(10).plot(kind='bar')
plt.title("Top Job Roles")
plt.xticks(rotation=45)
plt.show()
```

```python
top_loc = df['company_location'].value_counts().head(10).index
plt.figure()
sns.boxplot(x='company_location', y='salary_in_usd',
            data=df[df['company_location'].isin(top_loc)])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=df)
plt.title("Remote Work vs Salary")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='company_size', y='salary_in_usd', data=df)
plt.title("Company Size vs Salary")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
